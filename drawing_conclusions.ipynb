### Q2: What are the characteristics of SmartWay vehicles in 2008?

To find the characteristics of smartway vehicle, I plotted bars that compare smartway and non-smartway vehicles in mean of each category from 2008 data. 

```python
df_08_yes = df_08.query("smartway == 'yes'").mean()
df_08_no = df_08.query("smartway == 'no'").mean()

heights1 = [df_08_yes.air_pollution_score, df_08_no.air_pollution_score]
labels = ['smart_vehicle','non_smart_vehicle']
plt.bar([1,2], heights1, tick_label = labels)
plt.title('Average air pollution score of smart vehicle and non-smart vehicle in 2008')
plt.xlabel('Vehicle type')
plt.ylabel('Air pollution score');
```

Above graph shows that in air pollution score, smart vehicle > non_smart_vehicle.

```python
heights1 = [df_08_yes.city_mpg, df_08_no.city_mpg]
labels = ['smart_vehicle','non_smart_vehicle']
plt.bar([1,2], heights1, tick_label = labels)
plt.title('Average city_mpg of smart vehicle and non-smart vehicle in 2008')
plt.xlabel('Vehicle type')
plt.ylabel('city_mpg');
```

Above graph shows that in city MPG, smart vehicle > non_smart_vehicle.
